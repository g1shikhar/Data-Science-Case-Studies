```python
data = pd.read_csv('Data Set.csv', index_col = 'date', parse_dates = True)
```

```python
data
```

```python
pd.date_range(data.index[0], data.index[-1], freq='M')
```

```python
# Making the data contiuos by filling missing data points with 0
data = data.reindex(pd.date_range(data.index[0], data.index[-1], freq='M')).fillna(0)
```

```python
data.index.freq='M'
```

```python
corr_matrix = data.corr()[['Response Variable']]
```

```python
# Subseeting the variables having the high correlation for prediction
col = corr_matrix[(corr_matrix['Response Variable'] > 0.8) | (corr_matrix['Response Variable'] < -0.8)].index.values
```

```python
col
```

```python
data_corr = data[['Feature 3', 'Feature 6', 'Feature 7', 'Feature 9', 'Feature 10',
       'Feature 11', 'Feature 12', 'Feature 13', 'Feature 17',
       'Response Variable']]
```

```python
# null values treatment with median
for i in col:
    data_corr[i].replace(np.nan, data_corr[i].mean(), inplace=True)
    print(data_corr[i].mean())
```

```python
data_corr
```

```python
train = data_corr[:144]
```

```python
test = data_corr[144:]
```

```python
train
```

```python
# checking for seasonality in the data
from statsmodels.tsa.seasonal import seasonal_decompose
results = seasonal_decompose(data_corr['Response Variable'])
```

```python
results.seasonal.plot(figsize=(12,5));
# seasonality =12 (annula seasonality) as seen from graph
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# Since +ve acf at index 1 so it has ar(p) component
plot_acf(train['Response Variable'], lags=10);
```

```python
# sharp decilen in after the index 1 suggest ar(p)=1, ma(q)=0
plot_pacf(train['Response Variable'], lags=10);
```

```python
from statsmodels.tsa.stattools import adfuller 
```

```python
from statsmodels.tsa.statespace.tools import diff
# diff(train['Response Variable'], k_diff=1)
```

```python
#checking for stationarity to decide the value of d in (p,d,q)
dftest = adfuller(diff(train['Response Variable'], k_diff=1))
dfout = pd.Series(dftest[:4], index = ['ADF test statistics', 'p - value', '# lags used', '# of obsrvations'])
for key, val in dftest[4].items():
    dfout[f'critical value({key})'] = val
dfout
# d=1
```

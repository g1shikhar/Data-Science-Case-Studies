```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
order_data_engineered = pd.read_csv('data/Order_Data_Preprocess.csv', )
label_data=pd.read_csv("data/machine_learning_challenge_labeled_data.csv")
```

## Feature Engineering / Data Transformation

1. payment_id, platform_id, transmission_id - Nominal Categorical features <br>
     one hot encoding followed by taking sum at the customer level
<br>
<br>
2. order_hour, customer_order_rank - ordinal categorical features <br>
     min,  max, mode at customer level
<br>
<br>
3. voucher_amount  delivery_fee  amount_paid - continuos features <br>
   sum total and mean at customer level  
<br>
<br>
4. order_date - Various DateTime feature as shown in code<br>
    




```python
print("\n")
print("----------- One hot encoding for Nominal Categorical features'-------------")
print("\n")

for c in ['payment_id', 'platform_id', 'transmission_id']:

    order_data_engineered= pd.get_dummies(order_data_engineered, columns=[c])

print(order_data_engineered.columns)
print("\n")
print(order_data_engineered.shape)
```

```python
# Dropping 'restaurant_id', 'city_id' as transforming these nominal variables can 
# lead to sparse dataset given too many categories

order_data_engineered.drop(['restaurant_id', 'city_id'], axis=1, inplace=True)
```

```python
print("\n")
print("----------- Creating Date time features for order date-------------")
print("\n")

order_data_engineered['order_date'] = pd.to_datetime(order_data_engineered['order_date'],format='%Y-%m-%d')

order_data_engineered['year']=order_data_engineered['order_date'].dt.year 
order_data_engineered['month']=order_data_engineered['order_date'].dt.month 

order_data_engineered['dayofweek_num']=order_data_engineered['order_date'].dt.dayofweek  
order_data_engineered['dayofweek_name']=order_data_engineered['order_date'].dt.day_name()

order_data_engineered['weekend'] = np.where(order_data_engineered['dayofweek_name'].isin(['Sunday','Saturday']),1,0)

# once we have weekend column, we can drop of 'dayofweek_name'
order_data_engineered.drop('dayofweek_name', axis=1, inplace=True)

print(order_data_engineered.columns)
order_data_engineered.head()

print("\n")
print("----------- One hot encoding for Year, Month, dayofweek_num features-------------")
print("\n")

for c in ['year', 'month', 'dayofweek_num']:

    order_data_engineered= pd.get_dummies(order_data_engineered, columns=[c])
    
# once we have date time features, we can drop of 'order_date'
order_data_engineered.drop('order_date', axis=1, inplace=True)

print(order_data_engineered.columns)
order_data_engineered.head()
```

```python
print("\n")
print("----------- Aggregating Data at customet level with transformations-------------")
print("\n")

aggregation = dict.fromkeys(('payment_id_1491', 'payment_id_1523', 'payment_id_1619',
       'payment_id_1779', 'payment_id_1811', 'platform_id_525',
       'platform_id_22167', 'platform_id_22263', 'platform_id_22295',
       'platform_id_29463', 'platform_id_29495', 'platform_id_29751',
       'platform_id_29815', 'platform_id_30135', 'platform_id_30199',
       'platform_id_30231', 'platform_id_30359', 'platform_id_30391',
       'platform_id_30423', 'transmission_id_212', 'transmission_id_1988',
       'transmission_id_2020', 'transmission_id_4196', 'transmission_id_4228',
       'transmission_id_4260', 'transmission_id_4324', 'transmission_id_4356',
       'transmission_id_4996', 'transmission_id_21124', 'weekend', 'year_2015',
       'year_2016', 'year_2017', 'month_1', 'month_2', 'month_3', 'month_4',
       'month_5', 'month_6', 'month_7', 'month_8', 'month_9', 'month_10',
       'month_11', 'month_12', 'dayofweek_num_0', 'dayofweek_num_1',
       'dayofweek_num_2', 'dayofweek_num_3', 'dayofweek_num_4',
       'dayofweek_num_5', 'dayofweek_num_6'), 'sum')

aggregation2 = {'order_hour':['max', 'min'], 'customer_order_rank': ['max', 'min'],
                   'voucher_amount':['sum', 'mean'], 'delivery_fee':['sum', 'mean'], 'amount_paid':['sum', 'mean']}

aggregation.update(aggregation2)
customer_date_engineered = order_data_engineered.groupby('customer_id').agg(aggregation).reset_index()

# flatten MultiIndex columns
customer_date_engineered.columns = ['_'.join(col) for col in customer_date_engineered.columns.values]
customer_date_engineered.rename(columns = {list(customer_date_engineered)[0]: 'customer_id'}, inplace = True)

print(customer_date_engineered.shape)
print("\n")
print(customer_date_engineered.columns)
print("\n")
customer_date_engineered.head()
```

```python
print("\n")
print("----------- Merging Data with Labels-------------")
print("\n")

customer_date_engineered =  customer_date_engineered.merge(label_data, on = 'customer_id')

print(customer_date_engineered.shape)
print("\n")
print(customer_date_engineered.columns)
print("\n")

print("\n")
print("----------- Saving Engineered Dataset as 'customer_level_features'-------------")
print("\n")

customer_date_engineered.to_csv('data/customer_level_features.csv')
```
